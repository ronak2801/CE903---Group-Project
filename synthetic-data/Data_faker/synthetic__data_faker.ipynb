```python
from faker import Faker
import pandas as pd
import numpy as np
```

```python
fake=Faker()
```

```python

customer_id=[]
customer_name=[]
product_id_purchased=[]
purchase_date_time=[]
invoice_id=[]


for x in range(1000):

  invoice_id.append(x+1)
  
  #customer_id.append(x)
  
  name=(fake.name())

  for y in range(np.random.randint(1,10)):

    purchase_date_time.append(fake.date_time_between('-1y','now'))

    product_id_purchased.append(np.random.randint(1,210))

    customer_id.append(x+1)

    customer_name.append(name)


    
```

```python
df=pd.DataFrame(zip(customer_id,customer_name,purchase_date_time,product_id_purchased,invoice_id),columns=['customer_id','customer_name','purchase_date_time','product_id_purchased','invoice_id'])
df
```
