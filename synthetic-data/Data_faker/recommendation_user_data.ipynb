```python
import pandas as pd
df = pd.read_excel('E:\\Essex\\Project_903\\data_set\\_user_data.xlsx')
df
```

```python
df.info()
```

```python
df1=df.copy()

```

```python
#remove these columns and normalize the columns
df2=df1["Weekly_Budget"]
df4=df1.drop(["Categories","Allergies","Weekly_Budget"],axis=1) # file with customer_id, customer_name
data=df1.drop(["Weekly_Budget"],axis=1) # df1 is without weekly budget column
#define function
def normalize(data):
    normalize_data=(data-data.min())/(data.max()-data.min())
    return normalize_data
#call the function to normalize the columns 
normalized_data_frame=normalize(df2)
normalized_data_frame.head()
df4
```

```python

#df3=df1.copy()
df3=data.drop(["customer_id","customer_name"],axis=1)
df3
```

```python
# for calegorical data use one_hot encoding
one_hot=pd.get_dummies(df3[["Categories","Allergies"]])
one_hot.head()
```

```python
user_data=pd.concat([normalized_data_frame,one_hot],axis=1)
new_user_data=pd.concat([df4,user_data],axis=1)
new_user_data.head()
```

```python
new_user_data.to_excel("E:\\Essex\\Project_903\\data_set\\normalized_user_data.xlsx",index=False)
```
